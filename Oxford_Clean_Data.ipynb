```python
# Import Dependencies 
import pandas as pd
import numpy as np
```

```python
#Read CSV file
oxford_data = pd.read_csv("Resources/oxford_latest.csv", low_memory=False)
oxford_data.head(10)
```

```python
# Create new dataframe with `CountryName` as index
oxford_data = oxford_data.set_index("CountryName")
oxford_data
```

```python
# Create new dataframe with the 12 countries of interest\n",
countries = oxford_data.loc[["Brazil", "Canada", "China", "Germany", "Italy", "Japan", \
                             "New Zealand", "South Korea", "Sweden", "United Kingdom", "United States"]]

oxford_df = pd.DataFrame(countries, columns = ["Date", "C1_School closing", "C2_Workplace closing", "C6_Stay at home requirements", \
                                               "H6_Facial Coverings", "ConfirmedCases", "ConfirmedDeaths"])
oxford_df.reset_index()
```

```python
#Save new DataFrame to a csv file
oxford_df.to_csv("output_data/oxford_clean.csv", index=False)
```
